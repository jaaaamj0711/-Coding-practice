# Matplotlib 기본 사용법

## Pyplot 소개
- matplotlib.pyplot MATLAB과 비슷하게 명령어 스타일로 동작하는 파이썬 시각화 모듈입니다.

## 기본 그래프

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.ylabel('y-label')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
```

## 스타일 지정
: x, y 값 인자에 대해 그래프 선의 색상과 형태를 지정하는 포맷 문자열을 세번째 인자에 입력할 수 있습니다.

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], 'rs--')
plt.show()
```

- 첫번째값은 **색상**, 두번째값은 **마커** 스타일, 세번째값은 **선 스타일**을 의미합니다.
- 기본 포맷 문자열은 ‘b-‘이고 파란색(‘blue’)와 선(‘-‘)을 의미합니다.
- ‘rs-’는 빨간색(‘red’)와 사각형(‘ㅁ’), 선 모양('--')을 의미합니다.





### 색상

색상을 지정하는 방법은 색 이름이나 약자 또는 # 문자로 시작되는 RGB코드를 사용합니다. 색상코드는 다음 [사이트](http://matplotlib.org/examples/color/named_colors.html)에서 참고합니다.



### 마커

마커의 종류는 다음과 같습니다. 

| 마크 문자 | 의미 | 
| :--- | :---: |
| . | point marker| 
| o | circle marker | 
| * |star marker|
|+|  plus marker|

- 이 외에도 다양한 마커들이 존재합니다.

### 선 스타일

| 마크 문자 | 의미 | 
| :--- | :---: |
| - | solid line style| 
| -- | dashed line style | 
| -.|dash-dot line style|
|:|  dotted line style|

## 여러 개의 그래프 그리기

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0., 5., 0.2)


plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

- 다음과 같이 하나의 그래프로 여러값을 표현할 수도 있습니다.
